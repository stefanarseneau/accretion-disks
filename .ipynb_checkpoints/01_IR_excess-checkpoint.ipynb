```python
from astropy.io import fits
from astropy.table import Table, join, unique

from astropy.coordinates import SkyCoord
import astropy.units as u

from astropy.utils.metadata import merge

import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt


from multiprocessing import set_start_method

plt.style.use('stefan.mplstyle')
```

```python
catalog = Table.read('data/all_wds.csv')
references = Table.read('data/table_irsa_catalog_search_results.csv')

# Stars from Gaia selected to be within 100pc of the Sun
allstars = pd.read_csv('data/allstars_wise.csv')
#allstars = allstars.dropna(subset=['phot_g_mean_mag_01', 'phot_bp_mean_mag_01', 'phot_rp_mean_mag_01', 'parallax_01'])
allstars = Table.from_pandas(allstars)
```

```python
type(allstars['phot_g_mean_mag_01'][0])
```

```python
plt.figure(figsize = (10,3))
plt.hist(references['dist_x'], bins = 30)
plt.axvline(x = 2, c = 'k')
plt.xlabel('Distance [arcsec]')
plt.ylabel('Count')

references = references[references['dist_x'] < 2]
```

```python
plt.figure(figsize = (10,3))
plt.hist(allstars['dist_x'], bins = 30)
plt.axvline(x = 3.7, c = 'k')
plt.yscale('log')
plt.xlabel('Distance [arcsec]')
plt.ylabel('Count')

allstars = allstars[allstars['dist_x'] < 3.7]
```

```python
catalog = join(left=catalog, right=references, keys_left='CATALOGID', keys_right='catalogid_01', join_type="inner")
```

```python
plt.figure(figsize=(10,3))
plt.hist(catalog['PARALLAX'])
```

```python
#allstars = allstars[allstars['parallax_01'] < 50]
```

```python
plt.figure(figsize=(10,3))
plt.hist(allstars['parallax_01'])
```

```python
allstars['abs_g'] = np.array(allstars['phot_g_mean_mag_01'], dtype=float) + 5 * np.log10(allstars['parallax_01'] / 100)
allstars['abs_bp'] = allstars['phot_bp_mean_mag_01'] + 5 * np.log10(allstars['parallax_01'] / 100)
allstars['abs_rp'] = allstars['phot_rp_mean_mag_01'] + 5 * np.log10(allstars['parallax_01'] / 100)

allstars['abs_w1'] = allstars['w1mpro'] + 5 * np.log10(allstars['parallax_01'] / 100)
allstars['abs_w2'] = allstars['w2mpro'] + 5 * np.log10(allstars['parallax_01'] / 100)
allstars['abs_w3'] = allstars['w3mpro'] + 5 * np.log10(allstars['parallax_01'] / 100)

catalog['abs_g'] = catalog['GAIA_G'] + 5 * np.log10(catalog['PARALLAX'] / 100)
catalog['abs_bp'] = catalog['GAIA_BP'] + 5 * np.log10(catalog['PARALLAX'] / 100)
catalog['abs_rp'] = catalog['GAIA_RP'] + 5 * np.log10(catalog['PARALLAX'] / 100)

catalog['abs_w1'] = catalog['w1mpro'] + 5 * np.log10(catalog['PARALLAX'] / 100)
catalog['abs_w2'] = catalog['w2mpro'] + 5 * np.log10(catalog['PARALLAX'] / 100)
catalog['abs_w3'] = catalog['w3mpro'] + 5 * np.log10(catalog['PARALLAX'] / 100)
#catalog['abs_w4'] = catalog['w4mpro'] + 5 * np.log10(catalog['PARALLAX'] / 100)
```

```python
plt.figure(figsize=(18,7))

plt.subplot(1,2,1)

plt.scatter(allstars['abs_bp'] - allstars['abs_rp'], allstars['abs_g'], c = 'k', alpha = 0.01)
#plt.scatter(catalog['abs_bp'] - catalog['abs_rp'], catalog['abs_g'], c = 'red', alpha = 0.3)

plt.xlabel(r'$G_{BP} - G_{RP}$')
plt.ylabel('G')
plt.gca().invert_yaxis()

plt.subplot(1,2,2)

plt.scatter(allstars['abs_bp'] - allstars['abs_rp'], allstars['abs_w1'] - allstars['abs_w2'], c = 'k', alpha = 0.01)
#plt.scatter(catalog['abs_bp'] - catalog['abs_rp'], catalog['abs_w1'] - catalog['abs_w2'], c = 'red', alpha = 0.3)

plt.ylim(-2,2)

plt.xlabel(r'$G_{BP} - G_{RP}$')
plt.ylabel('W1 - W2')
```
